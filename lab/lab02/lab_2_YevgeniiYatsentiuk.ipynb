```python
import time

start_time = time.time()

# Сортування методом вставляння
def insertion_sort(nums):
    # починаємо з другого елемента
    for j in range(1, len(nums)):
        key = nums[j]
        # Зберігаємо посилання на індекс першого елемента
        i = j - 1
        # Елементи відсортованого сегмента переміщуємо вперед, якщо вони більше елемента вставки
        while i >= 0 and nums[i] > key:
            nums[i + 1] = nums[i]
            i -= 1
        # Вставляємо елемент
        nums[i + 1] = key

# Перевірка
random_list_of_numbers = [0, -5, 120, 18, 21, 4]
insertion_sort(random_list_of_numbers)
print(random_list_of_numbers)
print("--- %s seconds ---" % (time.time() - start_time))
```

**1. Що таке pip і conda та чим вони відрізняються між собою?**

**pip** — це стандартний менеджер пакетів Python.
встановлює пакети лише з PyPI
працює тільки з пакетами Python
не керує віртуальними середовищами (лише через venv)

**conda** — це менеджер пакетів і віртуальних середовищ.
    встановлює пакети з Anaconda та conda-forge
    працює не лише з Python (можна ставити C-бібліотеки, OpenCV, NumPy)
    має вбудовану систему створення та керування віртуальними середовищами
    краще встановлює важкі наукові пакети

**Головна різниця:**
pip = менеджер пакетів Python
conda = менеджер пакетів + менеджер середовищ (ширше і потужніше)

**4. Процедура сортування масиву методом бульбашки (Python)**

```python
print ('Сортування від меньшого до більшого')
my_list = [8, 10, 5, 2, 7]
swapped = True

while swapped:
    swapped = False
    for i in range(len(my_list) - 1):
        if my_list[i] > my_list[i + 1]:
            swapped = True
            my_list[i], my_list[i + 1] = my_list[1 + i], my_list[i]
            print(my_list)

my_list = [8, 10, 5, 2, 7]
swapped = True
print("Сортування від більшого до меньшого")
while swapped:
    swapped = False
    for i in range(len(my_list) - 1):
        if my_list[i] < my_list[i + 1]:
            swapped = True
            my_list[i], my_list[i + 1] = my_list[1 + i], my_list[i]
            print(my_list)
```
