**Обчислення факторіала за допомогою циклу `for`**

```python
def factorial_for(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
    
factorial_for(5)
```

**Обчислення факторіала рекурсивно**

```python
def factorial_recursive(n):
    if n == 0:
        return 1
    return n * factorial_recursive(n - 1)

factorial_recursive(5)
```

**Обчислення n-го числа Фібоначчі рекурсивно**

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n in (1, 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

fibonacci(10)
```

**Сортування злиттям (Merge Sort)**

```python
def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

def mergesort(arr):
    if len(arr) < 2:
        return arr
    mid = len(arr) // 2
    return merge(mergesort(arr[:mid]), mergesort(arr[mid:]))

mergesort([6, 5, 3, 1, 8, 7, 2, 4])
```

   Символ $\Omega$ (Омега велике) вказує на нижню межу складності алгоритму. Це означає, що алгоритм виконуватиметься *не швидше*, ніж ця оцінка (найкращий або мінімально необхідний час).

2. **$F(N) = N^3 + 7N^2 - 14N$. Записати в $O()$.**
   Відповідь: **$O(N^3)$**.
   (Беремо доданок з найбільшим степенем).

3. **$F(N) = 1.01^N + N^{10}$. Записати в $O()$.**
   Відповідь: **$O(1.01^N)$**.
   (Показникова функція, навіть з малою основою, зростає швидше за будь-який поліном при великих $N$).

4. **$F(N) = N^{1.3} + 10 \log_2 N$. Записати в $O()$.**
   Відповідь: **$O(N^{1.3})$**.
   (Степенева функція зростає швидше за логарифм).
