**Оцінка асимптотичної складності**

* **Insert (push):** $O(1)$ — додавання в кінець списку Python виконується за константний час.

## 2. Реалізація структури Черга
Використовуємо код класу Queue, де додавання відбувається у початок списку (insert 0), а видалення — з кінця.

```python
def print_n(queue_obj, n):

    if queue_obj.isEmpty():
        print("Черга пуста")
        return

    current_size = queue_obj.size()
    if n > current_size:
        n = current_size
    
    print(f"Виведення перших {n} елементів черги:")
    for i in range(1, n + 1):
        print(queue_obj.items[-i])


q2 = Queue()
for i in ["Перший", "Другий", "Третій", "Четвертий"]:
    q2.enqueue(i)

print_n(q2, 3)
```

**Оцінка асимптотичної складності**

Для наведеної реалізації (на базі list):

* **Insert (enqueue):** $O(n)$ — тому що використовується метод `insert(0, item)`, який зсуває всі існуючі елементи в пам'яті. Це "важка" операція для списку.
* **Delete (dequeue):** $O(1)$ — видалення з кінця (`pop()`) виконується миттєво.
* **Search:** $O(n)$ — пошук елемента вимагає проходу по списку.

## Відповіді на контрольні питання

**1. Що таке стек і які операції можна виконувати зі стеком?**
Стек — це структура даних, що працює за принципом LIFO (Last In — First Out), тобто останній доданий елемент видаляється першим.
Основні операції:
- **push(item)** — додавання елемента на вершину стека
- **pop()** — видалення й повернення верхнього елемента
- **peek()** — перегляд верхнього елемента без видалення
- **isEmpty()** — перевірка, чи стек порожній
- **size()** — кількість елементів у стеці

**2. Яка основна відмінність між стеком та чергою?**
Основна відмінність у порядку обробки елементів:
- **Стек** працює як стопка книг: LIFO (останній доданий забирається першим).
- **Черга** працює як черга в магазині: FIFO (First In, First Out — перший доданий обслуговується першим).

**3. Як ви можете реалізувати стек за допомогою масиву і за допомогою зв'язаного списку?**
* **Масив (список Python):** Елементи зберігаються підряд у пам'яті. Плюс: швидкий доступ. Мінус: якщо масив статичний, його треба розширювати (копіювати).
* **Зв'язаний список:** Кожен елемент має посилання на попередній. Плюс: легко додавати безліч елементів без повного копіювання. Мінус: займає більше пам'яті на зберігання посилань.
