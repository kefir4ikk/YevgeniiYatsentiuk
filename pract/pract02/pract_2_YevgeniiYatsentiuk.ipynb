## –ü—Ä–∞–∫—Ç–∏—á–Ω–∞ —Ä–æ–±–æ—Ç–∞ ‚Ññ 2 –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤. –Ü–Ω—à—ñ –Ω–æ—Ç–∞—Ü—ñ—ó
## –ú–µ—Ç–∞: –Ω–∞–±—É—Ç–∏ –ø—Ä–∞–∫—Ç–∏—á–Ω–∏—Ö –Ω–∞–≤–∏—á–æ–∫ —É —Ä–æ–∑–≤‚Äô—è–∑–∞–Ω–Ω—ñ –∑–∞–¥–∞—á –Ω–∞ –æ—Ü—ñ–Ω–∫—É –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–æ—ó —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤ —É Œ©, Œò, ùúä, ùúÉ, ùúî-–Ω–æ—Ç–∞—Ü—ñ—è—Ö.
### –í–∏–∫–æ–Ω–∞–≤: –Ø—Ü–µ–Ω—Ç—é–∫ –Ñ–≤–≥–µ–Ω—ñ–π, –≥—Ä—É–ø–∞: –ö–Ü-24-1

–í–∏–±—ñ—Ä –≤–∞—Ä—ñ–∞–Ω—Ç–∞
–ó–≥—ñ–¥–Ω–æ –∑ –ø—Ä–∞–≤–∏–ª–æ–º $n, n+5$, –¥–µ $n=23$ (–Ω–æ–º–µ—Ä —É —Å–ø–∏—Å–∫—É):
1.  –ó–∞–¥–∞—á–∞ 1: $23 - 15 = 8$ 
2.  –ó–∞–¥–∞—á–∞ 2: $23 + 5 = 28 \rightarrow 28 - 15 = 13$.

## –ú–µ—Ç–æ–¥ –º–µ–∂

–î–ª—è –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è –ø–æ—Ä—è–¥–∫—É –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ–π $f(n)$ —Ç–∞ $g(n)$ –æ–±—á–∏—Å–ª—é—î—Ç—å—Å—è –º–µ–∂–∞:
$$L = \lim_{n \to \infty} \frac{f(n)}{g(n)}$$

–Ü–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü—ñ—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É:
* –Ø–∫—â–æ $L = 0$, —Ç–æ $f(n)$ –∑—Ä–æ—Å—Ç–∞—î –ø–æ–≤—ñ–ª—å–Ω—ñ—à–µ –∑–∞ $g(n)$ ($f(n) = o(g(n)) \implies f(n) = O(g(n))$).
* –Ø–∫—â–æ $L = c > 0$ (–∫–æ–Ω—Å—Ç–∞–Ω—Ç–∞), —Ç–æ —Ñ—É–Ω–∫—Ü—ñ—ó –∑—Ä–æ—Å—Ç–∞—é—Ç—å –æ–¥–Ω–∞–∫–æ–≤–æ —à–≤–∏–¥–∫–æ ($f(n) = \Theta(g(n))$).
* –Ø–∫—â–æ $L = \infty$, —Ç–æ $f(n)$ –∑—Ä–æ—Å—Ç–∞—î —à–≤–∏–¥—à–µ ($f(n) = \omega(g(n))$).

## –ó–∞–≤–¥–∞–Ω–Ω—è 1 (‚Ññ 8 —É —Å–ø–∏—Å–∫—É)

**–£–º–æ–≤–∞:** –ú–∞—î–º–æ —Ñ—É–Ω–∫—Ü—ñ—ó $f(n) = 2n^3 - 4n^2 + 8$ (–≤—ñ–¥–Ω–æ–≤–ª–µ–Ω–æ –∑ –∫–æ–Ω—Ç–µ–∫—Å—Ç—É —è–∫ –ø–æ–ª—ñ–Ω–æ–º 3-–≥–æ —Å—Ç—É–ø–µ–Ω—è) —Ç–∞ $g(n) = n^4$. –ü–æ–∫–∞–∑–∞—Ç–∏, —â–æ $f(n) = O(g(n))$, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –º–µ—Ç–æ–¥ –º–µ–∂.

**–†–æ–∑–≤'—è–∑–∞–Ω–Ω—è:**
–ó–Ω–∞–π–¥–µ–º–æ –º–µ–∂—É –≤—ñ–¥–Ω–æ—à–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ–π –ø—Ä–∏ $n \to \infty$:

$$\lim_{n \to \infty} \frac{f(n)}{g(n)} = \lim_{n \to \infty} \frac{2n^3 - 4n^2 + 8}{n^4}$$

–†–æ–∑–¥—ñ–ª–∏–º–æ —á–∏—Å–µ–ª—å–Ω–∏–∫ –ø–æ—á–ª–µ–Ω–Ω–æ –Ω–∞ –∑–Ω–∞–º–µ–Ω–Ω–∏–∫:

$$= \lim_{n \to \infty} \left( \frac{2n^3}{n^4} - \frac{4n^2}{n^4} + \frac{8}{n^4} \right)$$
$$= \lim_{n \to \infty} \left( \frac{2}{n} - \frac{4}{n^2} + \frac{8}{n^4} \right)$$

–û—Å–∫—ñ–ª—å–∫–∏ –ø—Ä–∏ $n \to \infty$ –≤—Å—ñ –¥–æ–¥–∞–Ω–∫–∏ –ø—Ä—è–º—É—é—Ç—å –¥–æ 0:
$$= 0 - 0 + 0 = 0$$

**–í–∏—Å–Ω–æ–≤–æ–∫:**
–û—Å–∫—ñ–ª—å–∫–∏ –º–µ–∂–∞ –¥–æ—Ä—ñ–≤–Ω—é—î $0$, —Ü–µ –æ–∑–Ω–∞—á–∞—î, —â–æ $f(n)$ –∑—Ä–æ—Å—Ç–∞—î —Å—Ç—Ä–æ–≥–æ –ø–æ–≤—ñ–ª—å–Ω—ñ—à–µ, –Ω—ñ–∂ $g(n)$.
–ó–≥—ñ–¥–Ω–æ –∑ –ø—Ä–∞–≤–∏–ª–æ–º –º–µ—Ç–æ–¥—É –º–µ–∂ (–¥–∏–≤. —Ä–∏—Å. 1.9 —É –º–µ—Ç–æ–¥–∏—á—Ü—ñ), —è–∫—â–æ –º–µ–∂–∞ –¥–æ—Ä—ñ–≤–Ω—é—î 0, —Ç–æ $f(n) = o(g(n))$, —â–æ –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ –æ–∑–Ω–∞—á–∞—î $f(n) = O(g(n))$.

–¢–≤–µ—Ä–¥–∂–µ–Ω–Ω—è –¥–æ–≤–µ–¥–µ–Ω–æ.

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(1, 10, 100)

# –§—É–Ω–∫—Ü—ñ—ó
f_n = 2 * n**3 - 4 * n**2 + 8
g_n = n**4

plt.figure(figsize=(10, 6))
plt.plot(n, f_n, label=r'$f(n) = 2n^3 - 4n^2 + 8$')
plt.plot(n, g_n, label=r'$g(n) = n^4$', linestyle='--')

plt.title('–ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ–π (–ó–∞–≤–¥–∞–Ω–Ω—è 8)')
plt.xlabel('n')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–Ω—è')
plt.legend()
plt.grid(True)
plt.show()
```

## –ó–∞–≤–¥–∞–Ω–Ω—è 2 (‚Ññ 13 —É —Å–ø–∏—Å–∫—É)

**–£–º–æ–≤–∞:** –ó–∞–¥–∞–Ω–æ —Ñ—É–Ω–∫—Ü—ñ—ó $f(n) = n^3 - n^2 + 2$ —Ç–∞ $g(n) = n^4$. –ü–æ–∫–∞–∑–∞—Ç–∏, —â–æ $f(n) = O(g(n))$, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –º–µ—Ç–æ–¥ –º–µ–∂.

**–†–æ–∑–≤'—è–∑–∞–Ω–Ω—è:**
–û–±—á–∏—Å–ª–∏–º–æ –º–µ–∂—É:

$$\lim_{n \to \infty} \frac{f(n)}{g(n)} = \lim_{n \to \infty} \frac{n^3 - n^2 + 2}{n^4}$$

–í–∏–Ω–µ—Å–µ–º–æ —Å—Ç–∞—Ä—à–∏–π —Å—Ç–µ–ø—ñ–Ω—å $n^4$ —É –∑–Ω–∞–º–µ–Ω–Ω–∏–∫—É –∞–±–æ –ø—Ä–æ—Å—Ç–æ –ø–æ–¥—ñ–ª–∏–º–æ –ø–æ—á–ª–µ–Ω–Ω–æ:

$$= \lim_{n \to \infty} \left( \frac{n^3}{n^4} - \frac{n^2}{n^4} + \frac{2}{n^4} \right)$$
$$= \lim_{n \to \infty} \left( \frac{1}{n} - \frac{1}{n^2} + \frac{2}{n^4} \right)$$

–ü—ñ–¥—Å—Ç–∞–≤–∏–º–æ $n \to \infty$:
$$= 0 - 0 + 0 = 0$$

**–í–∏—Å–Ω–æ–≤–æ–∫:**
–û—Å–∫—ñ–ª—å–∫–∏ –º–µ–∂–∞ –¥–æ—Ä—ñ–≤–Ω—é—î $0$, —Ü–µ –ø—ñ–¥—Ç–≤–µ—Ä–¥–∂—É—î, —â–æ –ø–æ—Ä—è–¥–æ–∫ –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è $f(n)$ –º–µ–Ω—à–∏–π –∑–∞ –ø–æ—Ä—è–¥–æ–∫ $g(n)$.
–û—Ç–∂–µ, $f(n) = O(g(n))$ —î –≤—ñ—Ä–Ω–∏–º.

```python
n = np.linspace(1, 10, 100)

f_n_2 = n**3 - n**2 + 2
g_n_2 = n**4

plt.figure(figsize=(10, 6))
plt.plot(n, f_n_2, label=r'$f(n) = n^3 - n^2 + 2$')
plt.plot(n, g_n_2, label=r'$g(n) = n^4$', linestyle='--')

plt.title('–ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ–π (–ó–∞–≤–¥–∞–Ω–Ω—è 13)')
plt.xlabel('n')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–Ω—è')
plt.legend()
plt.grid(True)
plt.show()
```

## –í—ñ–¥–ø–æ–≤—ñ–¥—ñ –Ω–∞ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ñ –ø–∏—Ç–∞–Ω–Ω—è

**1. –©–æ —Ç–∞–∫–µ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É?**

–¶–µ –æ—Ü—ñ–Ω–∫–∞ –µ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—ñ –∞–ª–≥–æ—Ä–∏—Ç–º—É, —è–∫–∞ –ø–æ–∫–∞–∑—É—î, —è–∫ —à–≤–∏–¥–∫–æ –∑—Ä–æ—Å—Ç–∞—î —Ñ—É–Ω–∫—Ü—ñ—è $f(n)$ (—á–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∞–±–æ –æ–±—Å—è–≥ –ø–∞–º'—è—Ç—ñ) –ø—Ä–∏ –∑–±—ñ–ª—å—à–µ–Ω–Ω—ñ —Ä–æ–∑–º—ñ—Ä—É –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö $n$ –¥–æ –Ω–µ—Å–∫—ñ–Ω—á–µ–Ω–Ω–æ—Å—Ç—ñ ($n \to \infty$). –ù–∞—Å —Ü—ñ–∫–∞–≤–∏—Ç—å –Ω–µ —Ç–æ—á–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è —á–∞—Å—É –¥–ª—è –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–≥–æ $n$, –∞ —Å–∞–º–µ –ø–æ—Ä—è–¥–æ–∫ (—à–≤–∏–¥–∫—ñ—Å—Ç—å) –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó.

**2. –Ø–∫—ñ —ñ–Ω—à—ñ –Ω–æ—Ç–∞—Ü—ñ—ó, –∫—Ä—ñ–º $O$-–Ω–æ—Ç–∞—Ü—ñ—ó, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å—Å—è –¥–ª—è –≤–∏—Ä–∞–∂–µ–Ω–Ω—è –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–æ—ó —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ?**

–û–∫—Ä—ñ–º $O$-–Ω–æ—Ç–∞—Ü—ñ—ó (–≤–µ—Ä—Ö–Ω—è –º–µ–∂–∞), –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å—Å—è:
* $\Omega$ (–û–º–µ–≥–∞ –≤–µ–ª–∏–∫–∞) ‚Äî –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ –Ω–∏–∂–Ω—è –º–µ–∂–∞.
* $\Theta$ (–¢–µ—Ç–∞) ‚Äî –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ —Ç–æ—á–Ω–∞ –æ—Ü—ñ–Ω–∫–∞.
* $o$ (–æ –º–∞–ª–µ–Ω—å–∫–µ) ‚Äî —Å—É–≤–æ—Ä–∞ –≤–µ—Ä—Ö–Ω—è –º–µ–∂–∞.
* $\omega$ (–æ–º–µ–≥–∞ –º–∞–ª–µ–Ω—å–∫–∞) ‚Äî —Å—É–≤–æ—Ä–∞ –Ω–∏–∂–Ω—è –º–µ–∂–∞.

**3. –Ø–∫ –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Å–∏–º–≤–æ–ª—ñ–≤ $\Theta$ —ñ $\Omega$?**

* **–î–ª—è $\Omega$ (–Ω–∏–∂–Ω—è –º–µ–∂–∞):** –ü–æ—Ç—Ä—ñ–±–Ω–æ –∑–Ω–∞–π—Ç–∏ —Ç–∞–∫—ñ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–∏ $c > 0$ —Ç–∞ $n_0$, —â–æ–± –≤–∏–∫–æ–Ω—É–≤–∞–ª–∞—Å—è –Ω–µ—Ä—ñ–≤–Ω—ñ—Å—Ç—å $f(n) \ge c \cdot g(n)$ –¥–ª—è –≤—Å—ñ—Ö $n \ge n_0$. –¶–µ –æ–∑–Ω–∞—á–∞—î, —â–æ –∞–ª–≥–æ—Ä–∏—Ç–º –ø—Ä–∞—Ü—é—î *–ø—Ä–∏–Ω–∞–π–º–Ω—ñ* —Ç–∞–∫ —à–≤–∏–¥–∫–æ, —è–∫ $g(n)$ (–∞–±–æ –¥–æ–≤—à–µ).
* **–î–ª—è $\Theta$ (—Ç–æ—á–Ω–∞ –º–µ–∂–∞):** –ü–æ—Ç—Ä—ñ–±–Ω–æ –∑–Ω–∞–π—Ç–∏ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–∏ $c_1, c_2 > 0$ —Ç–∞ $n_0$, —â–æ–± –≤–∏–∫–æ–Ω—É–≤–∞–ª–∞—Å—è –ø–æ–¥–≤—ñ–π–Ω–∞ –Ω–µ—Ä—ñ–≤–Ω—ñ—Å—Ç—å $c_1 g(n) \le f(n) \le c_2 g(n)$ –¥–ª—è –≤—Å—ñ—Ö $n \ge n_0$. –¶–µ –æ–∑–Ω–∞—á–∞—î, —â–æ —Ñ—É–Ω–∫—Ü—ñ—è $f(n)$ –æ–±–º–µ–∂–µ–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—î—é $g(n)$ —ñ –∑–≤–µ—Ä—Ö—É, —ñ –∑–Ω–∏–∑—É.

**4. –Ø–∫–∞ —Ä—ñ–∑–Ω–∏—Ü—è –º—ñ–∂ $O$-–Ω–æ—Ç–∞—Ü—ñ—î—é, $\Theta$-–Ω–æ—Ç–∞—Ü—ñ—î—é —ñ $\Omega$-–Ω–æ—Ç–∞—Ü—ñ—î—é?**

* **$O(g(n))$**: –í–∏–∑–Ω–∞—á–∞—î **–≤–µ—Ä—Ö–Ω—é** –º–µ–∂—É. –§—É–Ω–∫—Ü—ñ—è —Ä–æ—Å—Ç–µ *–Ω–µ —à–≤–∏–¥—à–µ* –∑–∞ $g(n)$.
* **$\Omega(g(n))$**: –í–∏–∑–Ω–∞—á–∞—î **–Ω–∏–∂–Ω—é** –º–µ–∂—É. –§—É–Ω–∫—Ü—ñ—è —Ä–æ—Å—Ç–µ *–Ω–µ –ø–æ–≤—ñ–ª—å–Ω—ñ—à–µ* (–ø—Ä–∏–Ω–∞–π–º–Ω—ñ —Ç–∞–∫ —Å–∞–º–æ —à–≤–∏–¥–∫–æ) –∑–∞ $g(n)$.
* **$\Theta(g(n))$**: –í–∏–∑–Ω–∞—á–∞—î **—Ç–æ—á–Ω—É** –º–µ–∂—É. –§—É–Ω–∫—Ü—ñ—è —Ä–æ—Å—Ç–µ *–∑ —Ç—ñ—î—é –∂ —à–≤–∏–¥–∫—ñ—Å—Ç—é*, —â–æ –π $g(n)$ (–µ–∫–≤—ñ–≤–∞–ª–µ–Ω—Ç "–¥–æ—Ä—ñ–≤–Ω—é—î" –≤ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–æ–º—É —Å–µ–Ω—Å—ñ).

**5. –Ø–∫—ñ –æ—Å–Ω–æ–≤–Ω—ñ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ —ñ–Ω—à–∏—Ö –Ω–æ—Ç–∞—Ü—ñ–π, —Ç–∞–∫–∏—Ö —è–∫ $o$ (–º–∞–ª–µ–Ω—å–∫–∞ –æ) —Ç–∞ $\omega$ (–º–∞–ª–µ–Ω—å–∫–∞ –æ–º–µ–≥–∞)?**

–¶—ñ –Ω–æ—Ç–∞—Ü—ñ—ó (¬´–º–∞–ª—ñ¬ª –ª—ñ—Ç–µ—Ä–∏) —î –±—ñ–ª—å—à —Å—Ç—Ä–æ–≥–∏–º–∏ –≤–∞—Ä—ñ–∞–Ω—Ç–∞–º–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∏—Ö –æ—Ü—ñ–Ω–æ–∫ –ø–æ—Ä—ñ–≤–Ω—è–Ω–æ –∑ –≤–µ–ª–∏–∫–∏–º–∏ $O$ —Ç–∞ $\Omega$. –í–æ–Ω–∏ –≤–∫–∞–∑—É—é—Ç—å –Ω–∞ —Ç–µ, —â–æ —Ñ—É–Ω–∫—Ü—ñ—ó –∑—Ä–æ—Å—Ç–∞—é—Ç—å –∑ **—Ä—ñ–∑–Ω–æ—é** —à–≤–∏–¥–∫—ñ—Å—Ç—é (–Ω–µ—Ä—ñ–≤–Ω—ñ—Å—Ç—å —î —Å—Ç—Ä–æ–≥–æ—é).

* **$o(g(n))$ ‚Äî ¬´–æ –º–∞–ª–µ–Ω—å–∫–µ¬ª (–°—É–≤–æ—Ä–∞ –≤–µ—Ä—Ö–Ω—è –º–µ–∂–∞):**
    * **–°—É—Ç–Ω—ñ—Å—Ç—å:** –§—É–Ω–∫—Ü—ñ—è $f(n)$ –∑—Ä–æ—Å—Ç–∞—î **—Å—Ç—Ä–æ–≥–æ –ø–æ–≤—ñ–ª—å–Ω—ñ—à–µ**, –Ω—ñ–∂ $g(n)$. –¶–µ –æ–∑–Ω–∞—á–∞—î, —â–æ $f(n)$ —Å—Ç–∞—î –Ω–µ–∑–Ω–∞—á–Ω–æ—é –ø–æ—Ä—ñ–≤–Ω—è–Ω–æ –∑ $g(n)$ –ø—Ä–∏ –≤–µ–ª–∏–∫–∏—Ö $n$.
    * **–ú–∞—Ç–µ–º–∞—Ç–∏—á–Ω–µ –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è:** $f(n) = o(g(n))$, —è–∫—â–æ –¥–ª—è **–±—É–¥—å-—è–∫–æ—ó** –∫–æ–Ω—Å—Ç–∞–Ω—Ç–∏ $c > 0$ –∑–Ω–∞–π–¥–µ—Ç—å—Å—è —Ç–∞–∫–µ –∑–Ω–∞—á–µ–Ω–Ω—è $n_0$, —â–æ –¥–ª—è –≤—Å—ñ—Ö $n \ge n_0$ –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –Ω–µ—Ä—ñ–≤–Ω—ñ—Å—Ç—å $f(n) < c \cdot g(n)$.
    * **–í–∏–∑–Ω–∞—á–µ–Ω–Ω—è —á–µ—Ä–µ–∑ –≥—Ä–∞–Ω–∏—Ü—é:** $\lim_{n \to \infty} \frac{f(n)}{g(n)} = 0$.

* **$\omega(g(n))$ ‚Äî ¬´–æ–º–µ–≥–∞ –º–∞–ª–µ–Ω—å–∫–∞¬ª (–°—É–≤–æ—Ä–∞ –Ω–∏–∂–Ω—è –º–µ–∂–∞):**
    * **–°—É—Ç–Ω—ñ—Å—Ç—å:** –§—É–Ω–∫—Ü—ñ—è $f(n)$ –∑—Ä–æ—Å—Ç–∞—î **—Å—Ç—Ä–æ–≥–æ —à–≤–∏–¥—à–µ**, –Ω—ñ–∂ $g(n)$.
