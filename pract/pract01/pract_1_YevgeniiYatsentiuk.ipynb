### –í–∏–∫–æ–Ω–∞–≤: –Ø—Ü–µ–Ω—Ç—é–∫ –Ñ–≤–≥–µ–Ω—ñ–π, –≥—Ä—É–ø–∞: –ö–Ü-24-1

–ó–≥—ñ–¥–Ω–æ –∑ –ø—Ä–∞–≤–∏–ª–æ–º $n, n+5$, –¥–µ $n=23$ (–Ω–æ–º–µ—Ä —É —Å–ø–∏—Å–∫—É):
1.  –ó–∞–¥–∞—á–∞ 1: $23 - 15 = 8$

## –ó–∞–≤–¥–∞–Ω–Ω—è 1 (‚Ññ 8 —É —Å–ø–∏—Å–∫—É)

**–£–º–æ–≤–∞:** –î–æ–≤–µ—Å—Ç–∏, —â–æ $f(n) = 2n^2 + 10n + 3 = O(n^2)$.

**–†–æ–∑–≤'—è–∑–∞–Ω–Ω—è:**
–ó–≥—ñ–¥–Ω–æ –∑ –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è–º, $f(n) = O(g(n))$, —è–∫—â–æ —ñ—Å–Ω—É—é—Ç—å –∫–æ–Ω—Å—Ç–∞–Ω—Ç–∏ $c > 0$ —Ç–∞ $n_0 \ge 1$, —Ç–∞–∫—ñ —â–æ $f(n) \le c \cdot g(n)$ –¥–ª—è –≤—Å—ñ—Ö $n \ge n_0$.

–û—Ü—ñ–Ω–∏–º–æ —Ñ—É–Ω–∫—Ü—ñ—é –∑–≤–µ—Ä—Ö—É –¥–ª—è $n \ge 1$:
$$2n^2 + 10n + 3 \le 2n^2 + 10n^2 + 3n^2$$
(–º–∏ –∑–∞–º—ñ–Ω–∏–ª–∏ $n$ —Ç–∞ $1$ –Ω–∞ $n^2$, –æ—Å–∫—ñ–ª—å–∫–∏ –ø—Ä–∏ $n \ge 1$ —Ü–µ –∑–±—ñ–ª—å—à—É—î –∞–±–æ –∑–∞–ª–∏—à–∞—î –∑–Ω–∞—á–µ–Ω–Ω—è –Ω–µ–∑–º—ñ–Ω–Ω–∏–º).

$$2n^2 + 10n^2 + 3n^2 = 15n^2$$

–û—Ç–∂–µ:
$$2n^2 + 10n + 3 \le 15n^2$$

**–†–µ–∑—É–ª—å—Ç–∞—Ç:**
–ú–∏ –∑–Ω–∞–π—à–ª–∏ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–∏:
* $c = 15$
* $n_0 = 1$

–û—Å–∫—ñ–ª—å–∫–∏ –Ω–µ—Ä—ñ–≤–Ω—ñ—Å—Ç—å –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è, —Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è –¥–æ–≤–µ–¥–µ–Ω–æ.

```python
import matplotlib.pyplot as plt
import numpy as np

# –î—ñ–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω—å n
n = np.linspace(1, 20, 100)

# –§—É–Ω–∫—Ü—ñ—ó
f_n = 2 * n**2 + 10 * n + 3
g_n_c = 15 * n**2  # c * g(n), –¥–µ c = 15

# –ü–æ–±—É–¥–æ–≤–∞ –≥—Ä–∞—Ñ—ñ–∫–∞
plt.figure(figsize=(10, 6))
plt.plot(n, f_n, label=r'$f(n) = 2n^2 + 10n + 3$', linewidth=2)
plt.plot(n, g_n_c, label=r'$c \cdot g(n) = 15n^2$', linestyle='--', linewidth=2)

plt.title('–í—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è –¥–æ–≤–µ–¥–µ–Ω–Ω—è –∑–∞–≤–¥–∞–Ω–Ω—è ‚Ññ8')
plt.xlabel('n (—Ä–æ–∑–º—ñ—Ä –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö)')
plt.ylabel('–ö—ñ–ª—å–∫—ñ—Å—Ç—å –æ–ø–µ—Ä–∞—Ü—ñ–π')
plt.legend()
plt.grid(True)
plt.show()
```

## –ó–∞–≤–¥–∞–Ω–Ω—è 2 (‚Ññ 13 —É —Å–ø–∏—Å–∫—É)

**–£–º–æ–≤–∞:** –ó–∞–¥–∞–Ω–æ —Ñ—É–Ω–∫—Ü—ñ—ó $f(n) = 100n^3 + 8$ —Ç–∞ $g(n) = n^3$. –î–æ–≤–µ–¥—ñ—Ç—å, —â–æ $f(n) = O(g(n))$.

**–†–æ–∑–≤'—è–∑–∞–Ω–Ω—è:**
–ü–æ—Ç—Ä—ñ–±–Ω–æ –∑–Ω–∞–π—Ç–∏ $c$ —Ç–∞ $n_0$, —â–æ–± –≤–∏–∫–æ–Ω—É–≤–∞–ª–∞—Å—å –Ω–µ—Ä—ñ–≤–Ω—ñ—Å—Ç—å:
$$100n^3 + 8 \le c \cdot n^3$$

–û—Ü—ñ–Ω–∏–º–æ –≤–∏—Ä–∞–∑ –¥–ª—è $n \ge 1$:
1. –ß–ª–µ–Ω $100n^3$ –∑–∞–ª–∏—à–∞—î–º–æ –±–µ–∑ –∑–º—ñ–Ω.
2. –ö–æ–Ω—Å—Ç–∞–Ω—Ç—É $8$ –∑–∞–º—ñ–Ω—é—î–º–æ –Ω–∞ $8n^3$ (–æ—Å–∫—ñ–ª—å–∫–∏ $1 \le n^3$ –ø—Ä–∏ $n \ge 1$, —Ç–æ $8 \le 8n^3$).

–û—Ç—Ä–∏–º—É—î–º–æ:
$$100n^3 + 8 \le 100n^3 + 8n^3 = 108n^3$$

–¢–∞–∫–∏–º —á–∏–Ω–æ–º, –¥–ª—è $c = 108$ —Ç–∞ $n_0 = 1$ —É–º–æ–≤–∞ $f(n) \le c \cdot g(n)$ –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è.

**–†–µ–∑—É–ª—å—Ç–∞—Ç:**
–î–æ–≤–µ–¥–µ–Ω–æ, —â–æ $f(n) = O(n^3)$ –∑ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–æ—é $c=108$.

```python
# –î—ñ–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω—å n
n = np.linspace(1, 10, 100)

# –§—É–Ω–∫—Ü—ñ—ó
f_n_2 = 100 * n**3 + 8
g_n_c_2 = 108 * n**3  # c * g(n), –¥–µ c = 108

# –ü–æ–±—É–¥–æ–≤–∞ –≥—Ä–∞—Ñ—ñ–∫–∞
plt.figure(figsize=(10, 6))
plt.plot(n, f_n_2, label=r'$f(n) = 100n^3 + 8$', linewidth=2)
plt.plot(n, g_n_c_2, label=r'$c \cdot g(n) = 108n^3$', linestyle='--', linewidth=2)

plt.title('–í—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è –¥–æ–≤–µ–¥–µ–Ω–Ω—è –∑–∞–≤–¥–∞–Ω–Ω—è ‚Ññ13')
plt.xlabel('n')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó')
plt.legend()
plt.grid(True)
plt.show()
```

## –í—ñ–¥–ø–æ–≤—ñ–¥—ñ –Ω–∞ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ñ –ø–∏—Ç–∞–Ω–Ω—è

**1. –©–æ —Ç–∞–∫–µ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É?**

**2. –Ø–∫–∏–º —á–∏–Ω–æ–º –≤–∏–∑–Ω–∞—á–∞—î—Ç—å—Å—è $O$-–Ω–æ—Ç–∞—Ü—ñ—è —ñ —è–∫–∞ —ó—ó —Å—É—Ç–Ω—ñ—Å—Ç—å?**

* **–í–∏–∑–Ω–∞—á–µ–Ω–Ω—è:** $f(n) = O(g(n))$, —è–∫—â–æ —ñ—Å–Ω—É—é—Ç—å —Ç–∞–∫—ñ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–∏ $c > 0$ —Ç–∞ $n_0$, —â–æ –¥–ª—è –≤—Å—ñ—Ö $n \ge n_0$ –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –Ω–µ—Ä—ñ–≤–Ω—ñ—Å—Ç—å $f(n) \le c \cdot g(n)$.
* **–°—É—Ç–Ω—ñ—Å—Ç—å:** –í–æ–Ω–∞ –æ–ø–∏—Å—É—î –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É **–≤–µ—Ä—Ö–Ω—é –º–µ–∂—É** —á–∞—Å—É –≤–∏–∫–æ–Ω–∞–Ω–Ω—è. –¶–µ –æ–∑–Ω–∞—á–∞—î, —â–æ —Ñ—É–Ω–∫—Ü—ñ—è $f(n)$ —Ä–æ—Å—Ç–µ ¬´–Ω–µ —à–≤–∏–¥—à–µ¬ª, –Ω—ñ–∂ $g(n)$ (–∞–±–æ —â–æ–Ω–∞–π–±—ñ–ª—å—à–µ —Ç–∞–∫ —Å–∞–º–æ —à–≤–∏–¥–∫–æ). –ú–µ—Ç–∞ –≤–≤–µ–¥–µ–Ω–Ω—è $n_0$ ‚Äî —ñ–≥–Ω–æ—Ä—É–≤–∞—Ç–∏ –ø–æ–≤–µ–¥—ñ–Ω–∫—É —Ñ—É–Ω–∫—Ü—ñ—ó –Ω–∞ –º–∞–ª–∏—Ö –∑–Ω–∞—á–µ–Ω–Ω—è—Ö $n$.

**3. –Ø–∫—ñ –æ—Å–Ω–æ–≤–Ω—ñ –ø—Ä–∞–≤–∏–ª–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è $O$-–Ω–æ—Ç–∞—Ü—ñ—ó –ø—Ä–∏ –∞–Ω–∞–ª—ñ–∑—ñ –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤?**

–Ü—Å–Ω—É—é—Ç—å –¥–≤–∞ –æ—Å–Ω–æ–≤–Ω—ñ –ø—Ä–∞–≤–∏–ª–∞ —Å–ø—Ä–æ—â–µ–Ω–Ω—è:
1.  **–ü—Ä–∞–≤–∏–ª–æ –¥–æ–º—ñ–Ω—É–≤–∞–Ω–Ω—è (—Å—É–º–∏):** –Ø–∫—â–æ —Ñ—É–Ω–∫—Ü—ñ—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∞ —Å—É–º–æ—é –∫—ñ–ª—å–∫–æ—Ö —á–ª–µ–Ω—ñ–≤, —Ç–æ –ø–æ—Ä—è–¥–æ–∫ –≤–∏–∑–Ω–∞—á–∞—î —Ç–æ–π —á–ª–µ–Ω, —è–∫–∏–π —Ä–æ—Å—Ç–µ –Ω–∞–π—à–≤–∏–¥—à–µ. –í—ñ–Ω –ø–æ—á–∏–Ω–∞—î –¥–æ–º—ñ–Ω—É–≤–∞—Ç–∏ –ø—Ä–∏ $n \to \infty$, —Ä–æ–±–ª—è—á–∏ –≤–Ω–µ—Å–æ–∫ —ñ–Ω—à–∏—Ö –Ω–µ–∑–Ω–∞—á–Ω–∏–º.

**4. –©–æ –æ–∑–Ω–∞—á–∞—é—Ç—å –≤–∏—Ä–∞–∑–∏ $O(1)$, $O(n)$, $O(n^2)$ –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç—ñ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–æ—ó —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ?**
