## Відповіді на контрольні запитання

**1. Що таке кодування Гафмена та як воно працює?**

**Кодування Гафмена** — це алгоритм стиснення даних без втрат, який використовує коди змінної довжини.
**Принцип роботи:** Алгоритм присвоює коротші бінарні коди символам, які зустрічаються в повідомленні найчастіше, і довші коди — символам, які зустрічаються рідко. Завдяки цьому загальний розмір закодованого файлу зменшується. Це кодування є **префіксним**, тобто жоден код символу не є початком коду іншого символу, що дозволяє однозначно декодувати потік бітів без використання роздільників.

**2. Як визначається оптимальний двійковий код Гафмена для стиснення даних?**

Оптимальний код визначається через побудову **дерева Гафмена**:
1.  Визначаються частоти (ймовірності) входження кожного символу.
2.  Символи сортуються за зростанням частоти.
3.  На кожному кроці об’єднуються два вузли з найменшими частотами, утворюючи новий вузол із сумарною частотою.
4.  Процес повторюється, доки не залишиться один кореневий вузол.
5.  Гілкам дерева присвоюються біти (зазвичай 0 — лівій, 1 — правій). Шлях від кореня до листа з символом утворює його унікальний двійковий код.

**3. Які переваги має кодування Гафмена над іншими методами стиснення даних?**

* **Ефективність для нерівномірних даних:** Дає високий коефіцієнт стиснення, якщо частоти символів суттєво відрізняються (наприклад, у звичайному тексті).
* **Простота реалізації:** Алгоритм є зрозумілим і вимагає порівняно мало обчислювальних ресурсів.
* **Відсутність втрат:** Гарантує повне відновлення вихідної інформації (на відміну від JPEG чи MP3).
* **Префіксна властивість:** Дозволяє декодувати дані "на льоту" без необхідності використовувати спеціальні маркери кінця символу.

**4. Як відбувається декодування даних, закодованих за допомогою кодування Гафмена?**

Декодування здійснюється шляхом обходу дерева Гафмена від кореня до листів:
1.  Зчитується біт із закодованого потоку.
2.  Якщо біт дорівнює 0, переходимо лівою гілкою, якщо 1 — правою.
3.  Рух продовжується, доки не буде досягнуто "листа" дерева (кінцевого вузла).
4.  Символ, що знаходиться в листі, записується у вихідний файл, а процес повертається до кореня дерева для зчитування наступного біта.

**5. Які є можливі недоліки кодування Гафмена?**

* **Необхідність передачі словника:** Разом із закодованими даними потрібно передавати або саме дерево, або таблицю частот, що трохи збільшує обсяг файлу (особливо для коротких повідомлень).
* **Статичність (у класичному варіанті):** Алгоритм вимагає двох проходів по даних (перший — для підрахунку частот, другий — для кодування). Якщо статистика джерела змінюється в процесі, ефективність падає.
* **Чутливість до помилок:** Зміна навіть одного біта в потоці може призвести до неправильного декодування всієї подальшої інформації (розсинхронізація).

**6. Для чого використовується побудова дерева в кодуванні Гафмена?**

Побудова дерева необхідна для:
1.  **Гарантії префіксності:** Структура дерева автоматично забезпечує умову, що жоден код не є префіксом іншого (оскільки символи знаходяться тільки в листах, а не у внутрішніх вузлах).
2.  **Візуалізації та формування кодів:** Дерево дозволяє чітко визначити бітову послідовність для кожного символу.
3.  **Процесу декодування:** Дерево слугує "картою" для відновлення початкових даних із бітового потоку.
